```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
# some of the colors I will use in plotting

colors = ['#FF8000', '#0673F5', '#09B235', '#FF3336', '#14FFDF', '#8D14FF', '#FFE800', '#49FF00']
```

```python
df = pd.read_csv('model_results.csv')
```

```python
df
```

```python
columns = df.columns[1:]
```

```python
plt.figure(figsize=(13,8))

sns.distplot(df[['Middle_LR']], hist=False, color=colors[0])
sns.distplot(df[['Middle_SVC']], hist=False, color=colors[1])
sns.distplot(df[['Left_LR']], bins=5, hist=False, color=colors[2])
sns.distplot(df[['Left_SVC']], hist=False, color=colors[3])
sns.distplot(df[['Right_LR']], bins=5, hist=False, color=colors[4])
sns.distplot(df[['Right_SVC']], hist=False, color=colors[5])

plt.legend(labels=columns)
```

```python
plt.figure(figsize=(13,8))
sns.barplot(data=df, palette=colors)
```

```python
for i in range(len(columns)):
    plt.figure(figsize=(8,5))
    sns.barplot(x=df[columns[i]], y=df['relation_label'], palette=colors, ci=None)
```

```python
for i in range(len(columns)):
    plt.figure(figsize=(12,3))
    sns.scatterplot(y=df[columns[i]], x=df['relation_label'], hue=df['relation_label'], palette=colors, s=100)
    plt.legend(bbox_to_anchor=(1.01, 1.0), loc='upper left', borderaxespad=0)
```
